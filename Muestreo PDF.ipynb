```python
import random
import math
import collections
from matplotlib import pyplot as plt
```

# Funciones

```python
def d_normal(rand, it, b):
    list1 = []
    for i in range (0,it):
        sum = 0.0
        for i in range(0,b):
            sum += random.uniform(-rand, rand)
        list1.append(round(sum/2,2))
        
    dict1 = collections.Counter(list1)
    dict1 = dict(sorted(dict1.items(), key=lambda item: item[0])) 
    return dict1
```

```python
def d_triangular(rand, it):
    list1 = []
    for i in range (0,it):
        n = math.sqrt(6)/2*(random.uniform(-rand, rand) + random.uniform(-rand, rand))
        list1.append(round(n,2))
        
    dict1 = collections.Counter(list1)
    dict1 = dict(sorted(dict1.items(), key=lambda item: item[0])) 
    return dict1
```

```python
def plot(dist_dict):
    plt.bar(range(len(dist_dict)), list(dist_dict.values()), align='center')
    plt.xticks(range(len(dist_dict)), list(dist_dict.keys()))
    plt.show()
```

# Muestreo

## Distribución normal

Sumatoria de 12

```python
list1 = d_normal(1,10000,12)
list2 = d_normal(1,100000,12)
list3 = d_normal(1,10000000,12)
```

```python
plot(list1)
```

```python
plot(list2)
```

```python
plot(list3)
```
